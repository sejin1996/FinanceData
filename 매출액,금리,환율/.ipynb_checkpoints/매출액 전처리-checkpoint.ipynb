```python
list=['KB 금융','LG 생활건강','LG 화학','NAVER','SK 텔레콤','삼성물산','삼성바이오로직스','삼성전자','포스코 케미칼','한국전력','현대 모비스','현대차']
```

```python
df = pd.read_csv('data/KB 금융.csv',encoding='euc-kr',index_col='결산년도')
```

```python
df1=df[['당기순익']][:19]
```

```python
df1.to_csv('data/KB금융_매출액.csv',encoding='euc-kr')
```

```python
for i in list:
    df = pd.read_csv('data/'+i+'.csv',encoding='euc-kr',index_col='결산년도')
    df1=df[['당기순익']][:19]
    df1.to_csv('data/'+i+'_매출액.csv',encoding='euc-kr')
```

```python
df_5 = pd.read_csv('data/삼성바이오로직스_매출액.csv',encoding='euc-kr',index_col='결산년도')
df_6 = pd.read_csv('data/KB 금융_매출액.csv',encoding='euc-kr',index_col='결산년도')
pd.concat([df_5,df_6],axis=1,join='outer')
```

```python
df_merge=pd.DataFrame()
for i in list:
    df_i = pd.read_csv('data/'+i+'_매출액.csv',encoding='euc-kr',index_col='결산년도')
    df_i=df_i.rename(columns={'당기순익':i})
    try:
        df_merge = pd.concat([df_merge,df_i],axis=1)
    except:
        df_i.index=df_merge.index
        df_merge = pd.concat([df_merge,df_i],axis=1)
df_merge=df_merge[:-1]
```

```python
for i in df_merge.columns:
    for n in range(len(df_merge[i])):
        try:
            df_merge[i][n]=df_merge[i][n].replace(',','')
        except:
            pass
```

```python
df_merge=df_merge.astype(float)
```

```python
df_merge['NAVER']['21년03월(1Q)']=(df_merge['NAVER']['20년12월(4Q)']+2960)/2
# 네이버 21년1분기의 당기순익이 합병때문에 비정상적인 값을 가지고 있어 20년4분기와 21년2분기예상값의 평균으로 대체함
```

```python
df_merge=df_merge.sort_index()
df_merge=df_merge.fillna(value=0)
```

```python
df_merge.to_csv('당기순익_merge.csv',encoding='euc-kr')
```
